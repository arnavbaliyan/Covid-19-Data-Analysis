In the next 4 following charts we will create a choropleth map using plotly to look at the world view for the impact of the covid-19 virus and how many people were Confirmed cases, Active cases, Recovered cases, and also the Deaths for the countries in the world. When you hover over a country you will see the name of the country along with the statistics for that certian country based on what were are analyzing with that map

```python
figure = px.choropleth(country_wise, locations="Country/Region",
                    locationmode='country names', color="Confirmed",
                    hover_name="Country/Region",
                    color_continuous_scale="Peach",
                    title='Confirmed Cases For Countries')
figure.show()
```

```python
# Now we will create a choropleth graph for Countries with Active Cases
figure = px.choropleth(country_wise, locations="Country/Region",
                    locationmode='country names', color="Active",
                    hover_name="Country/Region",
                    color_continuous_scale="Peach",
                    title='Active Cases For Countries')
figure.show()
```

```python
# Now creating a choropleth graph for Recovered Cases for the Countries
figure = px.choropleth(country_wise, locations="Country/Region",
                    locationmode='country names', color="Recovered",
                    hover_name="Country/Region",
                    color_continuous_scale="Peach",
                    title='Recovered Cases For Countries')
figure.show()
```

```python
#Creating a choropleth map for the Deaths that occured in each country from Covid-19
figure = px.choropleth(country_wise, locations="Country/Region",
                    locationmode='country names', color="Deaths",
                    hover_name="Country/Region",
                    color_continuous_scale="Peach",
                    title='Death Cases For Countries')
figure.show()
```

Within in the next 4 charts I will be creating a bar chart using plotly to look at the top 15 countries with the follwoing variables of Deaths, Confirmed, Recovered, and Active cases for the top 15 countries within each variable to get some insight to see which country had a stronger recovery rate than others and which countries had a higher fatility number than others.

```python
top_15_recovered=country_wise.sort_values(by=['Deaths'],ascending=False)[0:15]
fig = px.bar(top_15_recovered, x='Country/Region', y='Deaths', template='plotly_dark')
fig.show()
```

```python
top_15_confirmed=country_wise.sort_values(by=['Confirmed'],ascending=False)[0:15]
fig = px.bar(top_15_confirmed, x='Country/Region', y='Confirmed',template='plotly_dark')
fig.show()
```

```python
top_15_recovered=country_wise.sort_values(by=['Recovered'],ascending=False)[0:15]
fig = px.bar(top_15_recovered, x='Country/Region', y='Recovered',template='plotly_dark')
fig.show()
```

```python
top_15_active=country_wise.sort_values(by=['Active'],ascending=False)[0:15]
fig = px.bar(top_15_active, x='Country/Region', y='Active',template='plotly_dark')
fig.show()
```

The next chart I will be creating a grouped bar chart also looking at the top 15 countries by their Confirmed cases because that is a relaible factor onto the number of overall cases within each country that was impacted with the covid-19 virus. This grouped bar chart will combine all 4 variables that we have previously anaylzed and can show us more clearly on how the countries were when they were contacted with the virus. After we made the chart we saw that US, Brazil, and India were the top 3 for Confirmed cases and for all countries the Recovery numbers were much higher than the Deaths which is a good sign.

```python
top_15 = country_wise.sort_values(by=['Confirmed'],ascending=False)[:15]
fig = go.Figure()
fig.add_trace(go.Bar(
                    x=top_15['Country/Region'],y=top_15['Confirmed'],name='Confirmed',marker_color='navy'
))
fig.add_trace(go.Bar(
                    x=top_15['Country/Region'],y=top_15['Recovered'],name='Recovered',marker_color='mediumaquamarine'
))
fig.add_trace(go.Bar(
                    x=top_15['Country/Region'],y=top_15['Active'],name='Active',marker_color='seagreen'
))
fig.add_trace(go.Bar(
                    x=top_15['Country/Region'],y=top_15['Deaths'],name='Deaths',marker_color='darkturquoise'
))

fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

For our next chart with using the day_wise data set, we will create a stacked scatter plot to compare the worldwide Confirmed, Recovered, and Deaths for all of the countries combined. My x-axis will be the dates so we can see how all of the varaibles increased over time and also see that the Confirmed cases and people Recovered start to turn exponential around the month of June 2020.

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=day_wise.Date, y=day_wise.Confirmed, mode="markers", name="Confirmed",fill='tozeroy'
))
fig.add_trace(go.Scatter(
    x=day_wise.Date, y=day_wise.Recovered, mode="markers", name="Recovered",fill='tozeroy'
))
fig.add_trace(go.Scatter(
    x=day_wise.Date, y=day_wise.Deaths, mode="markers", name="Deaths",fill='tozeroy'
))
fig.update_layout(title='Cases over time within the World',
                 xaxis_title='Dates',
                 yaxis_title='Population',
                 template = 'plotly_dark')

fig.show()
```

The following four lines of code will be used to inspect and analyze the data that is contained within the usa_county_wise.csv file. We can understand the variables in the data set and see how to optomize them for the analysis.

```python
usa_county_wise.head()
```

```python
usa_county_wise.describe()
```

```python
usa = usa_county_wise.groupby('Province_State').sum().reset_index()

usa_abbrev = {
    'Alabama': 'AL', 'Alaska': 'AK','American Samoa': 'AS', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
    'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'District of Columbia': 'DC', 'Florida': 'FL',
    'Georgia': 'GA', 'Guam': 'GU', 'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
    'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD', 'Massachusetts': 'MA',
    'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO', 'Montana': 'MT','Nebraska': 'NE',
    'Nevada': 'NV', 'New Hampshire': 'NH','New Jersey': 'NJ',
    'New Mexico': 'NM',  'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Northern Mariana Islands':'MP',
    'Ohio': 'OH', 'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 'Puerto Rico': 'PR',
    'Rhode Island': 'RI', 'South Carolina': 'SC', 'South Dakota': 'SD', 'Tennessee': 'TN',
    'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virgin Islands': 'VI', 'Virginia': 'VA', 'Washington': 'WA',
    'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
}
usa['State'] = usa['Province_State']
usa['State'] = usa['State'].replace(usa_abbrev)
usa['Death_rate'] = usa['Deaths'] / usa['Confirmed']
```

```python
fig = px.pie(usa, values='Deaths', names='Province_State', title='Death cases in the US')
fig.update_traces(textposition='inside', textinfo='label+percent')

fig.show()
```

```python
fig = px.pie(usa, values='Confirmed', names='Province_State', title='Confirmed cases in the US')
fig.update_traces(textposition='inside', textinfo='label+percent')

fig.show()
```

Based on the pie charts, we can see that the top three states with the most deaths in the country are New York, New Jersey, and Massachusetts. We can also see that the top three states with the most confirmed cases are New York, California, and New Jersey. We can imply that because of the high population in these condensed states cause the virus to spread quicker.

```python
figure = px.choropleth(usa, color='Confirmed', locations='State', locationmode="USA-states",
                    scope="usa", color_continuous_scale="dense", title='Confirmed cases in the USA')
figure.show()
```

```python
figure = px.choropleth(usa, color='Deaths', locations='State', locationmode="USA-states",
                    scope="usa", color_continuous_scale="dense", title='Death cases in the USA')
figure.show()
```

Up above are two choropleth graphs to show a geo graphical chart based on the cases for the Deaths and Confirmed cases in the United States of America. New York is shown to have the most confirmed cases with New Jersey and California coming in after that. From the charts it shows that the Midwest and upper West are shown to have less confirmed cases compare to the rest of the country. Moving towards the deaths, again New York has the most deaths and that the northern region of the country seem to have higher counts of casualities compared to the rest of the country
